```python
import matplotlib.pyplot as plt
import numpy as np

# Data
mean = {
    
    'Lens_L': 0,
    'Lens_R': 0,
    'Eye_L': 0.5,
    'Eye_R': 0.5,
    'Brain': 0.999,
    'Optic_Nerve_R': 0.94,
    'Optic_Nerve_L': 0.96,
    'Optic_Chiasm': 0.99,
    'Oral_Cavity': 1,
    'Lips': 1,
    'Mandible': 1,
    'Larynx': 1,
    'Pharynx': 1,
    'Spinal Canal': 1
}

std = {
    
    'Lens_L': 0,
    'Lens_R': 0,
    'Eye_L': 0.1,
    'Eye_R': 0.1,
    'Brain': 0.002,
    'Optic_Nerve_R': 0.03,
    'Optic_Nerve_L': 0.03,
    'Optic_Chiasm': 0.01,
    'Oral_Cavity': 0,
    'Lips': 0,
    'Mandible': 0,
    'Larynx': 0,
    'Pharynx': 0,
    'Spinal Canal': 0
}

labels = list(mean.keys())
means = list(mean.values())
stds = list(std.values())

x = np.arange(len(labels))  # the label locations
colors = ['orange'] * 4 + ['cornflowerblue'] * 4 + ['tomato'] * 6  # Define colors for the groups

fig, ax = plt.subplots(figsize=(12, 8))
ects1 = ax.barh(x, means, xerr=stds, capsize=5, color=colors, height=0.7)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Dice Score')
# ax.set_title('Means and Stan/dard Deviations')
ax.set_yticks(x)
ax.set_yticklabels(labels)
# ax.set_xticklabels(ax.get_xticks(), fontsize=14)
ax.invert_yaxis()  # Invert y axis to have the first item at the top

# Add legend
legend_labels = ['Anatomy Completely or Partially Removed', 'Anatomy on Same Slice as Crop', 'Anatomy Below Crop']
handles = [plt.Rectangle((0, 0), 1, 1, color=c, ec="k") for c in ['orange', 'cornflowerblue', 'tomato']]
ax.legend(handles, legend_labels, fontsize=14, title='Anatomy Categories', title_fontsize='16')

fig.tight_layout()

plt.show()
```
